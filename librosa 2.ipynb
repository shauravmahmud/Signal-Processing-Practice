# Audio signals & sampling

```python
import librosa
```

```python
y, sr = librosa.load(example, sr=None)
```

```python
print("Sample rate  :", sr)
print("Signal Length:", len(y))
print("Duration     :", len(y)/sr, "seconds")
```

## or

```python
Audio(y, rate=sr)
```

```python
print("Sample rate  :", sr)
print("Signal Length:", len(y))
print("Duration     :", len(y)/sr, "seconds")
```

```python
Audio(y, rate=sr/2)
```

```python
print("Sample rate  :", sr/2)
print("Signal Length:", len(y))
print("Duration     :", len(y)/(sr/2), "seconds")
```

```python
Audio(y, rate=sr*2)
```

```python
print("Sample rate  :", sr*2)
print("Signal Length:", len(y))
print("Duration     :", len(y)/(sr*2), "seconds")
```
